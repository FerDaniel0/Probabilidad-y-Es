```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
n = 3
p = 0.5
```

```python
x = np.arange(0,n+1)
pmf = binom.pmf(x,n,p)
```

```python
plt.bar(x,pmf)
plt.xlabel('Numero de soles')
plt.ylabel('Probabilidad')
plt.title('Distribucion Binomial: Numero de soles en 3 volados ')
plt.show()
```
