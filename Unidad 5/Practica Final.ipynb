```python
# Datos de entrada: Horas de estudio y calificaciones
horas_estudio = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
calificaciones = np.array([20, 25, 34, 30, 40, 31])

# Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Predecir la calificación para 8 horas de estudio
horas_prediccion = np.array([[8]])
calificacion_predicha = modelo.predict(horas_prediccion)

print("La calificación esperada para 8 horas de estudio es:", calificacion_predicha[0])
print("Valor de la intercepción (b):", modelo.intercept_)
print("Valor de la pendiente (m):", modelo.coef_[0])

# Crear el diagrama de dispersión con la línea de regresión
plt.figure(figsize=(6,5))
plt.scatter(horas_estudio, calificaciones, color='blue', label='Datos')
plt.plot(horas_estudio, modelo.predict(horas_estudio), color='red', label='Línea de regresión')
plt.scatter(horas_prediccion, calificacion_predicha, color='green', marker='x', s=100, label='Predicción (8 horas, {:.2f})'.format(calificacion_predicha[0]))
plt.title('Pronóstico de calificaciones según horas de estudio')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificación final')
plt.legend()
plt.grid(True)
plt.show()
```

# 1.¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?
El coeficiente de regresión pendiente indica cuánto se espera que aumente la calificación final de un estudiante por cada hora adicional de estudio. Es decir, si la pendiente es, por ejemplo, 3, esto significa que por cada hora extra de estudio, la calificación final aumenta en promedio 3 unidades.



# 2. ¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?
El coeficiente de correlación de Pearson r mide la fuerza y dirección de la relación lineal entre las horas de estudio y las calificaciones. Un valor cercano a +1 indica una relación lineal fuerte y positiva; es decir, a mayor tiempo de estudio, mayores calificaciones. Un valor cercano a 0 indica una relación débil o nula, y un valor cercano a -1 indica una relación fuerte y negativa.
