# Ejercicio 1
Una revision de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

La probabilidad de estudiantes que pesan mas de 68 y menos de 72 kg.
Cuantos estudiantes pesan mas de 68 y menos de 72 kg.

# Normal mente se utiliza
**norm.rvs**()para calcular numero aleatorios

**norm.pdf**()para calcular la dencidad de probabilidad

**norm.cdf**()Para calcular la probabilidad

**norm.ppf**()Para calcular el valor de la a partir de una probabilidad , es la funcion inversa de cdf()

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Generar Datos para la distribución normal

media = 70
desv_std = 4
num_datos = 500

#Generar datos de manera aleatoria
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50]
```

```python
# Graficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribucion normal')
plt.xlabel('Peso (kg)De los estudiantes')
plt.grid(True)
plt.show()
```

```python
# crear un rango de valores para crear uan densidad
x = np.linspace(datos - 3 * desv_std, media + 3 * desv_std, num_datos)
# calcular la PDF de la distribuvion normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std**2))
```

```python
#Para no imprimir los 500 valores solos los primeros 50
print("Imprimir los primero 50 valores de x",x[0:50],", Y los primero 50 valores de pdf",pdf[0:50])
```

```python
# visualizar la PDF
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='normal:(media:{};ds:{})'.format(np.round(media,2), np.round(desv_std,2)))
plt.title('Funcion de densidad de probabilidad (PDF)- Distribucion normal')
plt.xlabel('Peso (kg)De los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Graficar Hostograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')

plt.title('Distribucion normal')
plt.xlabel('Peso (kg)De los estudiantes')
plt.grid(True)
plt.show()
```
